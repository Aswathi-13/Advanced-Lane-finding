```python
# 1. CAMERA CALLIBRATION

def camCalib():
    images = glob.glob('camera_cal/calibration*.jpg') #reading all the calibration images
    objpoints =[] #3D points in real world space
    imgpoints =[] #2D points in image plane
    nx=9 # corners along x 
    ny=6 # corners along y
    objp = np.zeros((6*9,3),np.float32) #6*9 points, each with 3 columns
    objp[:,:2]=np.mgrid[0:9,0:6].T.reshape(-1,2) # x,y coordinates reshaping to 2D

    for fname in images:
        image = mpimg.imread(fname)
        #plt.imshow(image)
        gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #converting to gray scale
        ret,corners = cv2.findChessboardCorners(gray,(9,6),None) #function call to find the chess board corners
        if ret ==True:
            imgpoints.append(corners) # appending the corners to 2D array
            objpoints.append(objp) #appending the 3D array
            image = cv2.drawChessboardCorners(image,(9,6),corners,ret) # drawing the chess board corners
    #plt.imshow(image)
    [ret,mtx,dist,rvecs,tvecs]=cv2.calibrateCamera(objpoints,imgpoints,gray.shape[::-1],None,None) #function call to calibrate the camera
    return mtx,dist #Camera matrix, distortion coefficient

# 2. UNDISTORTING THE IMAGE
def undistortion(image,mtx,dist):
    udst = cv2.undistort(image,mtx,dist,None,mtx)
    return udst
    
```

```python
# 3. COLOR AND GRADIENT THRESHOLDING

#function definition for Sobel x filter
def sob_Threshold(img,thresh =(50,255)):
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) #converting to gray scale
    #APPLYING SOBEL FILTER
    abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    #converting the abs value to 8 bit
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel)) 
    #creating a binary threshold to select pixels based on gradient strength
    binary_sobelx = np.zeros_like(scaled_sobel)
    binary_sobelx[(scaled_sobel>=thresh[0]) & (scaled_sobel<=thresh[1])]=1    
    return binary_sobelx

#function definition to find the magnitude threshold
def mag_thresh(img, sobel_kernel=3, mag_thresh=(50,255)):
    #convert to gray scale
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    #the gradient in x and y separately
    sobx = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=sobel_kernel)
    soby = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=sobel_kernel)
    # magnitude calculation
    mag_sob = np.sqrt((sobx**2)+(soby**2))
    #converting the abs value to 8 bit
    scaled_sob = np.uint8(255*mag_sob/np.max(mag_sob))
    #creating a binary mask where mag thresholds are met
    binary_mag = np.zeros_like(scaled_sob)
    binary_mag[(scaled_sob>=mag_thresh[0])&(scaled_sob<mag_thresh[1])]=1
    return binary_mag

#function definition to find the directional threshold
def dir_threshold(img, sobel_kernel=3, thresh=(0.7, 1.25)):
    # Convert to grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    # gradient in x and y
    sobx = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=sobel_kernel)
    soby = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=sobel_kernel)
    # the absolute value of the x and y gradients
    abs_sobelx = np.absolute(sobx)
    abs_sobely = np.absolute(soby)
    # the direction of the gradient 
    grad_dir=np.arctan2(abs_sobely, abs_sobelx)
    # creating binary mask where direction thresholds are met
    binary_dir = np.zeros_like(grad_dir)
    binary_dir[(grad_dir>=thresh[0]) & (grad_dir<thresh[1])]=1
    return binary_dir

#function definition color space threshold
def hls_threshold(img, thresh=(170, 255)):
    #Convert RGB to HLS and threshold to binary image using S channel
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    s_channel = hls[:,:,2]
    # creating binary mask where thresholds are met
    binary_hls = np.zeros_like(s_channel)
    binary_hls[(s_channel > thresh[0]) & (s_channel <= thresh[1])] = 1
    return binary_hls

#function definition for combined binary thresholding
def com_binary(img):  
    #function call - sobel filter along x direction
    sob_binary = sob_Threshold(img,thresh =(50,255))
    #function call - mag thresholding
    mag_binary = mag_thresh(img, sobel_kernel=3, mag_thresh=(50,255))
    #function call - directional thresholding
    dir_binary = dir_threshold(img, sobel_kernel=3, thresh=(0.7, 1.25))
    #function call- colour space thresholding
    hls_binary = hls_threshold(img, thresh=(170, 255))
    #creating a mask where all the thresholds are met
    combined_binary = np.zeros_like(sob_binary)
    combined_binary[(sob_binary == 1) | (mag_binary  == 1) & (dir_binary  == 1) | (hls_binary ==1)] = 1
    return combined_binary
 
```

```python
#4. PERSPECTIVE TRANSFORM
def persTransform(img):
    #generating the source points (polygon) from the binary image
    src = np.float32([[190,720],[600,450],[690,450],[1100,700]])
    offset = 300
    img_size = (img.shape[1],img.shape[0])
    #creating destination coordinate points as the new corners of the image
    dest = np.float32([[offset, img_size[1]],[offset, 0],[img_size[0]-offset, 0],[img_size[0]-offset, img_size[1]]])
    #print(dest)
    #function call to get the birds eye view of the masked image
    M = cv2.getPerspectiveTransform(src, dest)
    #function call to convert back the birds eye view to real image
    invM = cv2.getPerspectiveTransform(dest, src)
    # Warp the image 
    warped = cv2.warpPerspective(img, M, img_size)
    return warped ,invM
```

```python
#5. FINDING THE LANE BOUNDARIES

#function definition to find the lane lines using sliding window
def find_lane_pixels(img):
    #generating a histogram to get the max peak
    histogram = np.sum(img[img.shape[0]//2:,:],axis =0)
    out_img = np.dstack((img,img,img))*255
    #splitting the histogram in the middle and finding the left lane and right lane base points respectively
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:])+midpoint
    
    #HYPERPARAMETERS
    nwindows = 9
    margin = 100
    minpix = 50
    window_height = np.int(img.shape[0]//nwindows)
    
    nonzero = img.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    
    leftx_current = leftx_base
    rightx_current = rightx_base
    
    left_lane_inds=[]
    right_lane_inds=[]
    
    for window in range(nwindows):
        #finding the sliding window coordinates of both lanes
        win_y_low = img.shape[0]-(window+1)*window_height
        win_y_high = img.shape[0]-(window*window_height)
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current +margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current +margin
        
        #finding the activation pixels in each window
        good_left_inds= ((nonzeroy>=win_y_low)&(nonzeroy<win_y_high)&(nonzerox>=win_xleft_low)&(nonzerox<=win_xleft_high)).nonzero()[0]
        good_right_inds= ((nonzeroy>win_y_low)&(nonzeroy<=win_y_high)&(nonzerox>win_xright_low)&(nonzerox<=win_xright_high)).nonzero()[0]
        
        #appending the lane indices
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        #to recenter the window if the length of indicies is more than req value
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
    try:
        #concatenating the indices
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        pass           
    
    #extracting the left and right lane pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds]
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    return leftx, lefty, rightx, righty

#function definition to draw the polynomial
def fit_poly(leftx, lefty, rightx, righty,img):
   
    #generating the xy values for plotting
    ploty = np.linspace(0,img.shape[0]-1,img.shape[0])
    out_img = np.dstack((img,img,img))*255
    #gernerating A,B,C coefficients of polynomial using polyfit function
    left_fit = np.polyfit(lefty,leftx,2)
    right_fit = np.polyfit(righty,rightx,2)
    
    try:
        #polynomial
        left_fitx = left_fit[0]*ploty**2+ left_fit[1]*ploty +left_fit[2]
        right_fitx = right_fit[0]*ploty**2+ right_fit[1]*ploty +right_fit[2]
    except TypeError:
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty
    return out_img,left_fit,right_fit,ploty

def plot_Poly(leftx, lefty, rightx, righty,img):
        
     #generating the xy values for plotting
    ploty = np.linspace(0,img.shape[0]-1,img.shape[0])
    out_img = np.dstack((img,img,img))*255
    #gernerating A,B,C coefficients of polynomial using polyfit function
    left_fit = np.polyfit(lefty,leftx,2)
    right_fit = np.polyfit(righty,rightx,2)
    
    try:
        #polynomial
        left_fitx = left_fit[0]*ploty**2+ left_fit[1]*ploty +left_fit[2]
        right_fitx = right_fit[0]*ploty**2+ right_fit[1]*ploty +right_fit[2]
    except TypeError:
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty
     
     ## Visualization ##
    # Colors in the left and right lane regions
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [0, 0, 255]

    # Plots the left and right polynomials on the lane lines
    plt.plot(left_fitx, ploty, color='yellow')
    plt.plot(right_fitx, ploty, color='yellow')
    plt.title('Polynomial Line', fontsize=25)
   
    return out_img,left_fit,right_fit,ploty
    
#function definition to find the lane from prior
def search_around_poly(img):
    
    margin = 100
    #defining the global variables
    global left_fit_before #left index value from sliding window
    global right_fit_before #right index value from sliding window
    
    #activation pixels along xy
    nonzero = img.nonzero()
    nonzerox = np.array(nonzero[1])
    nonzeroy = np.array(nonzero[0])
    #finding the activation pixels within +/- margin of the polynomial line
    left_lane_inds = ((nonzerox > (left_fit_before[0]*(nonzeroy**2) + left_fit_before[1]*nonzeroy + left_fit_before[2] - margin)) & (nonzerox < (left_fit_before[0]*(nonzeroy**2) + left_fit_before[1]*nonzeroy + left_fit_before[2] + margin)))
    right_lane_inds = ((nonzerox > (right_fit_before[0]*(nonzeroy**2) + right_fit_before[1]*nonzeroy + right_fit_before[2] - margin)) & (nonzerox < (right_fit_before[0]*(nonzeroy**2) + right_fit_before[1]*nonzeroy + right_fit_before[2] + margin)))
    
    # Again, extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]

    #out_img, left_fitx, right_fitx, ploty = fit_poly(img.shape, leftx, lefty, rightx, righty)
    return leftx, lefty, rightx, righty
```

```python
#6. RADIUS OF CURVATURE 

def measure_curvature_real(left_fit_cr,right_fit_cr,ploty):

    # Define conversions in x and y from pixels space to meters
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    
    # Define y-value where we want radius of curvature
    y_eval = np.max(ploty) * ym_per_pix
    
    # Implementing the calculation of R_curve (radius of curvature) #####
    left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])

    return left_curverad, right_curverad
```

```python
#7. POSITION OF THE VEHICLE IN METERS

def vehicle_position(left_fit,right_fit,undist_img):
    
    # meters per pixel in x dimension
    xm_per_pix = 3.7/700
    
    y_max = undist_img.shape[0]
       
    #creating the x polynomials at y max
    x_lt = left_fit[0]*y_max**2+left_fit[1]*y_max+left_fit[2]
    x_rt = right_fit[0]*y_max**2+right_fit[1]*y_max+right_fit[2]
    
    #finding the midpoint
    midpt = (x_lt+x_rt)//2
    #position of the vehicle
    veh_pos = ((undist_img.shape[1]//2) - midpt) * xm_per_pix 
    return veh_pos
```

```python
#8. WARP BACK TO REAL IMAGE

def warp_into_real(undist_img,left_fit,right_fit,ploty,left_curverad_m, right_curverad_m,veh_pos,invM):
    #warp_zero = np.zeros_like(undist_img).astype(np.uint8)
    #color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    color_warp = np.zeros((720, 1280, 3), dtype='uint8') 
    
    #generating the polygon
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]

    # recast the x and y points into usable format for cv2.fillPoly()
    leftlane = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    rightlane = np.array([np.flipud(np.transpose(np.vstack([right_fitx,ploty])))])
    lane_pts = np.hstack((leftlane, rightlane))
    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([lane_pts]), (0,255, 0))
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, invM, (undist_img.shape[1], undist_img.shape[0]))
    # Combine the result with the original image
    result = cv2.addWeighted(undist_img, 1, newwarp, 0.3, 0)
    
    # Annotate vehicle offset from center
    rad_curve = (left_curverad_m+right_curverad_m)/2
    str1 = 'Radius of curvature is '+str(rad_curve)[:7]
    result = cv2.putText(result, str1, (40,50), cv2.FONT_HERSHEY_COMPLEX_SMALL, 2, (255,255,255), 5, cv2.LINE_AA)

    str2 = 'vehicle offset is '+str(veh_pos)[:7] 
    result = cv2.putText(result, str2, (40,100), cv2.FONT_HERSHEY_COMPLEX_SMALL, 2, (255,255,255), 5, cv2.LINE_AA)
    
    return result
```

```python
#9. CHECKING ON VIDEO STREAM

#declaring the global parameters

global left_fit_hist 
left_fit_history = np.array([]) # to stack the old good left indices 

global right_fit_hist 
right_fit_history = np.array([]) # to stack the right good left indices 

#function definition to find lane
def lane_find(img):
    #function call to combined binary threshold
    binary_image = com_binary(img)
    #function call to perspective transform
    perc_image, invM = persTransform(binary_image)
    
    global left_fit_history 
    global right_fit_history
    global left_fit_before
    global right_fit_before
    
    # check if any lane points were known before
    if (len(left_fit_history) == 0):
        #if not, find lane pixels from histogram method and sliding window
        leftx, lefty, rightx, righty = find_lane_pixels(perc_image) #leftx, lefty, rightx, righty
        out_img_poly,left_fit,right_fit,ploty = fit_poly(leftx, lefty, rightx, righty, img)
        
        # Stack and store fit history
        left_fit_history = np.array(left_fit)
        new_left_fit = np.array(left_fit)
        left_fit_history = np.vstack([left_fit_history, new_left_fit])
        
        right_fit_history = np.array(right_fit)
        new_right_fit = np.array(right_fit)
        right_fit_history = np.vstack([right_fit_history, new_right_fit])
        
    #if the lane line points were know before
    else:
        #initialising the known point from previous method
        left_fit_before = [np.mean(left_fit_history[:,0]), np.mean(left_fit_history[:,1]), np.mean(left_fit_history[:,2])]
        right_fit_before = [np.mean(right_fit_history[:,0]), np.mean(right_fit_history[:,1]), np.mean(right_fit_history[:,2])]
        #function call to search from prior
        leftx, lefty, rightx, righty = search_around_poly(perc_image)
        
        #if the curve ran off the image
        if (len(lefty) == 0 or len(righty) == 0):
            leftx, lefty, rightx, righty, img_LanePixel_B4 = find_lane_pixels(perc_image)
        
        img_polyB4,left_fit,right_fit,ploty = fit_poly(leftx, lefty, rightx, righty, img)
        
        # Add new values to history
        new_left_fit = np.array(left_fit)
        left_fit_history = np.vstack([left_fit_history, new_left_fit])
        new_right_fit = np.array(right_fit)
        right_fit_history = np.vstack([right_fit_history, new_right_fit])
        # Remove old values from history
        if (len(left_fit_history) > 10):
            left_fit_history = np.delete(left_fit_history, 0,0)
            right_fit_history = np.delete(right_fit_history, 0,0)
    
    #function call to measure the radius                                   
    left_curverad, right_curverad =  measure_curvature_real(left_fit,right_fit,ploty)
    #function call to find the vehicle position                                 
    veh_pos = vehicle_position(left_fit,right_fit,img)
    #annoting the image and warping to real image
    result = warp_into_real(img,left_fit,right_fit,ploty,left_curverad, right_curverad,veh_pos,invM)#
    return result

video_output = 'project_video_output1.mp4'
clip1 = VideoFileClip("project_video.mp4")
output_clip = clip1.fl_image(lane_find)
%time output_clip.write_videofile(video_output, audio=False)
```
